```python
distance_table = {}
time_table = {}

origins = DictLocations.values()
destinations = DictLocations.values()

origin_idx = enumerate(origins)
destination_idx = enumerate(destinations)

print(origin_idx)


for (origin_idx, originX) in enumerate(origins):
    for (destination_idx, destinationX) in enumerate(destinations):
        matrix_distance = gmaps.distance_matrix(origins[origin_idx], destinations[destination_idx], mode="driving")
        error = matrix_distance['rows'][0]['elements'][0]['status']
        if error <> 'OK':
            print("error en los datos")
            print(origins[origin_idx], destinations[destination_idx])
        else: 
            distance_table[originX, destinationX] = 0.001 * (matrix_distance['rows'][0]['elements'][0]['distance']['value'])
        #time_table[originX, destinationX] = (1/3600)*(  matrix_distance['rows'][0]['elements'][0]['duration_in_traffic']['value'])
```
