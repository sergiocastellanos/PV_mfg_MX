```python
import pandas as pd
import time
```

```python
countiesDF = pd.read_csv(r'/counties.txt', header=-1) #make sure you place correctly the address where file is located
```

```python
finalListAlt=[]
nameListAlt = []
for i in countiesDF[0]:
    time.sleep(0.3)
    latlongMatrix = geocoder.osm(i).latlng
    print(latlongMatrix, i)
    finalListAlt.append(latlongMatrix)
    nameListAlt.append(i)

counties = pd.DataFrame(finalListAlt, nameListAlt)    
```

```python
countiesAlt = pd.DataFrame(finalListAlt, nameListAlt)
```

```python
countiesAlt.columns=['Latitude','Longitude']
countiesAlt
```

```python
countiesAlt.to_csv(r'/countiesLatLong.csv')
```
